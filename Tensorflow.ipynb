# Simple Linear Regression 

## Important relevant libraries 

```python
import matplotlib.pyplot as plt
from numpy import random as np_random
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

## Generate random input data to train on 

```python
observations = 1000

xs = np.random.uniform(low=-10,high=10,size = (observations,1))
zs = np.random.uniform(-10,10,(observations,1)) 

inputs = np.column_stack((xs,zs)) 
print (inputs.shape)
```

## Create targets we will aim at 

```python
noise = np.random.uniform(-1,1,(observations,1))
```

```python
targets = 2*xs - 3*zs + noise 
print(targets.shape)
```

## Plot the training data 

```python
targets = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot(xs, zs, targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')
ax.view_init(azim=0)
plt.show()
targets = targets.reshape(observations,1)
```


# Initialize Variables

```python
init_range = 0.1 
weights = np.random.uniform(-init_range, init_range, size = (2,1))
biases = np.random.uniform(-init_range, init_range, size = 1)
print(weights)
print(biases)
```


# Set a learning Rate

```python
learning_rate = 0.02
```


# Train the model 

```python
for i in range(100): 
    outputs = np.dot(inputs,weights) + biases 
    deltas = outputs - targets 
    loss = np.sum(deltas ** 2)/ 2 / observations 
  
    print(loss)
    
    delta_scaled = deltas / observations
    
    weights = weights - learning_rate * np.dot(inputs.T,delta_scaled)
    biases = biases - learning_rate * np.sum(delta_scaled)
```

```python
print(weights,biases)
```

# PLOT OUTPUT VS TARGETS

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
